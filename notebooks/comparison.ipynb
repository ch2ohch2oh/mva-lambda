```python
from root_pandas import read_root
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', size=12)
```

```python
data = read_root('../mc/exp55_run20_use_mva_mc.root', 'lambda')
```

```python
data.shape
```

```python
opts = {'histtype':'step', 'bins':100, 'range':[1.105, 1.130]}
fig = plt.figure(figsize=(8, 6))
plt.hist(data.query('goodBelleLambda >= 1 and p_pid_pk > 0.6 and p_pid_ppi > 0.6').M, label='goodBelleLambda >= 1', **opts)
plt.hist(data.query('goodBelleLambda >= 2 and p_pid_pk > 0.6 and p_pid_ppi > 0.6').M, label='goodBelleLambda >= 2', **opts)
plt.hist(data.query('mva > 0.97').M, label='mva > 0.97', **opts)
plt.legend()
plt.xlabel('Lambda0 mass (GeV)')
plt.ylabel('Counts per bin')
plt.title('goodBelleLambda compared with MVA lambda on MC')
plt.show()
```
